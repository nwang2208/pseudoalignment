```python
def TrieConstruction (patterns):
    trie = dict()
    trie[0] = dict()
    newNode = 1
    for word in patterns:
        currentNode = 0
        for i in range(0, len(word)):
            currentSymbol = word[i]
            if currentSymbol in trie[currentNode]:
                currentNode = trie[currentNode][currentSymbol]
            else:
                trie[newNode] = dict()
                trie[currentNode][currentSymbol] = newNode
                currentNode = newNode
                newNode += 1
    return trie
```

```python
def BWT(text):
    myList = []
    for i in range(0, len(text)):
        word = text[i:len(text)]
        if(i != 0):
            word += text[0:i]
        myList.append(word)
    myList.sort()

    lastCol = ""
    for el in myList:
        lastCol += el[len(el)-1]

    return(lastCol)

```

```python
def sortString(str):
    return ''.join(sorted(str))




def BWmatrix(text):
    tempWord = ""
    textDict = [['',0]] * len(text)
    for i in range(0, len(text)):
        if(tempWord.find(text[i]) == -1): #character has not shown up yet
            textDict[i] = [text[i],1]
        else: #duplicate character
            for x in range(len(textDict)-1,-1, -1):
                if(textDict[x][0] == text[i]):
                    textDict[i] = [text[i], textDict[x][1] + 1]
                    break
        tempWord += text[i]
    return(textDict)


def TransformBWT(text):
    lastColMatrix = BWmatrix(text)
    firstCol = sortString(text)
    firstColMatrix = BWmatrix(firstCol)

    outputArr = []
    index = lastColMatrix.index(['$',1])
    el = firstColMatrix[index]
    outputArr.append(el)

    for x in range(1, len(lastColMatrix)):
        index = lastColMatrix.index(el)
        el = firstColMatrix[index]
        outputArr.append(el)

    return(outputArr)





```
